```
# Start writing code here...
import pandas as pd
import numpy as np
ser = {
    'index':[0, 1, 2, 3],
    'data' :[100, 200, 300, 400],
    'name' : 'songs'
}

songs = {
    'index' : ['Sai', 'Khan', 'Hlawn', 'Paing'],
    'data' : [134, 150, 23, 35],
    'name' : 'counts'
}
def get(ser,idx):
    value_idx = ser['index'].index(idx)
    return ser['data'][value_idx]
```

```
test = get(ser, 1)
print(test)
#print(type(test))
get(songs, 'Paing')
```
