````
# Email address regex
# thetpaimyo@gmail.com
# thetpi@utycc.org 
# to escape special characters , use bashlash '\' in front of them 
test_email = re.compile(r'\w+@\w+\.[a-z]{2,5}')
text = "To email Thet Paing, try paing@phandeeyar.org or the personal address thetpai@gmail.com"
test_email.findall(text)
````

````
# Parentheses indicate groups to extract 
# "\w+"" to "[\w.]+" to better match for any alphanumeric character or a period 

email1 = re.compile(r'([\w.]+)@(\w+)\.([a-z]{3})')
text = "To email Thet Paing, try paing@phandeeyar.org or the personal address thetpai@gmail.com"
email1.findall(text)
````

````
# Extract/Match components as Python dictionary 
email4 = re.compile(r'(?P<username>[\w.]+)@(?P<domain>\w+)\.(?P<suf>[a-z]{3,5})')
mat = email4.match('thetpai@gmail.mmcom')
mat.groupdict()
````

````
x = 'From: Using the : character'
y = re.findall('^F.+:', x)
print(y)
````

## Regular Expression Methods 
- re.match() 
    - re.match() function of re in Python will search the regular expression pattern and return the first occurrence.
    - The Python RegEx Match method checks for a match only at the beginning of the string. So, if a match is found in the first line, it returns the match object. But if a match is found in some other line, the Python RegEx Match function returns null.
- re.search()
    - re.search() function will search the regular expression pattern and return the first occurrence.
- re.findall()
    - findall() module is used to search for “all” occurrences that match a given pattern.
