## Python Lambda Function (Dec 23)

1. well known as anonymous function , that doesn't have a name or function inside function. 
2. Anatomy of lambda function
    * **lambda** keyword tells Python for constructing a function
    * **signature** ,everything between lambda & : sign , means the function only takes how many arguments.(Example : variable(x))
    * return function after : sign

```
a = lambda y : y**2 
print(a(2))
b = lambda x,y,z: x*y*z
print(b(2,2,5))
print(type(b))
```

```
# Powerful technique (Nested function method - that I called by myself)
def test(x):
    return lambda a: a+x 
result = test(1000)
print(result(5))
```

```
# Another example 
unsorted = [('a', 8), ('c', 2), ('b', 5), ('d', 1)]
print(sorted(unsorted, key=lambda x:x[0]))
print(sorted(unsorted, key=lambda x:x[1]))

sum_fun = lambda x,y:x+y
print(sum_fun(10,40))
```
