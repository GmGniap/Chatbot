```
# Start writing code here...
import pandas as pd
```

```
staff = pd.DataFrame([
{'Name': 'U Mya', 'Role': 'Head', 'Loc':'Ygn'},
{'Name':'Daw Hla', 'Role': 'CEO', 'Loc': 'Mdy'},
{'Name': "Mg Mg", 'Role': 'Teacher', 'Loc': 'POL'}
]
)

student = pd.DataFrame([
    {'Name': 'Hla Hla', 'School': 'Business','Loc':'Ygn'},
    {'Name': 'Mg Mg', 'School': 'Law','Loc':'POL'}, 
    {'Name': 'Pai', 'School': 'Engineering', 'Loc': 'Mdy'}
])


staff.set_index('Name', inplace=True)
student.set_index('Name', inplace=True)
print(staff.head())
print(student.head())
```

```
pd.merge(staff, student, how='outer', left_index=True, right_index=True)
```

```
pd.merge(staff, student, how='inner', left_index=True, right_index=True)
```

```
pd.merge(staff, student, how='right', left_index=True, right_index=True)
```

```
staff_df = staff.reset_index()
student_df = student.reset_index()

pd.merge(student_df,  staff_df, how='inner', on=['Name','Loc'])
```

```
frame = [staff, student]
pd.concat(frame , axis = 1)
```
